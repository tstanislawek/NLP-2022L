```python
import os
from score_discopolo import score_discopolo, score_repetability
```

```python
scores_dict = {}
for samples_path in ['samples', 'MS/samples/disco_polo_gen', 
                     'samples/gpt2_validation_songs', 
                     'samples/disco_polo_gen_gpt2_tokenized',
                     'samples/disco_polo_gen_gpt2_validation',
                     'samples/disco_polo_gen']:
    for file in [os.path.join(samples_path, p) for p in os.listdir(samples_path) if p.startswith('sample')]:
        with open(file) as f: 
            try:
                song = f.read()
                scores_dict[song] = {"discopolo_score": score_discopolo(song), "repetability":  score_repetability(song)}
            except RuntimeError:
                pass #hard to fix right now, just skip 
```

```python
for k,v in sorted(scores_dict.items(), key=lambda item: item[1]["discopolo_score"])[:5]:
    print(k, v, '\n\n')
```

```python
for k,v in sorted(scores_dict.items(), key=lambda item: item[1]["discopolo_score"])[-5:]:
    print(k, v, '\n\n')
```

Song worst in terms of repetability: 

```python
for k,v in sorted(scores_dict.items(), key=lambda item: item[1]["repetability"])[:5]:
    print(k, v, '\n\n')
```

Song best in terms of repetability: 

```python
for k,v in sorted(scores_dict.items(), key=lambda item: item[1]["repetability"])[-5:]:
    print(k, v, '\n\n')
```

Joint score:

```python
min_rep = min([x['repetability'] for x in scores_dict.values()])
max_rep = max([x['repetability'] for x in scores_dict.values()])
min_dps = min([x['discopolo_score'] for x in scores_dict.values()])
max_dps = max([x['discopolo_score'] for x in scores_dict.values()])
for song, scores in scores_dict.items():
    scores_dict[song]['score'] = 0.5*((scores['repetability'] - min_rep)/(max_rep - min_rep) + (scores['discopolo_score'] - min_dps)/(max_dps - min_rep))
for k,v in sorted(scores_dict.items(), key=lambda item: item[1]["score"])[-5:]:
    print(k, v, '\n\n')
```
