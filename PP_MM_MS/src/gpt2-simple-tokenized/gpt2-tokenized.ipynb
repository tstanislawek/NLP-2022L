```python
import gpt_2_simple as gpt2
import os
from datetime import datetime 
import tensorflow as tf
from google.cloud import storage
os.chdir('..')
```

```python
model_name="124M"
checkpoint_dir = 'checkpoint'
file_name = "data/songs_data_tokenized.csv"
timestamp = datetime.now().strftime("%Y%m%d%H%M%S")
bucket_name = "poem-generator-checkpoints"
run_name='disco_polo_gen_gpt2_tokenized'
```

```python
if not os.path.exists(os.path.join('models/gpt2_tokenized', model_name)):
    gpt2.download_gpt2(model_dir="models/gtp2_tokenized", model_name=model_name)
```

```python
sess = gpt2.start_tf_sess() 
gpt2.finetune(sess,
              model_dir="models/gtp2_tokenized",
              dataset=file_name,
              model_name=model_name,
              reuse=False,
              steps=15000,
              run_name=run_name,
              print_every=10,
              sample_every=200,
              save_every=100, 
              only_train_transformer_layers=True,
              accumulate_gradients = 1,
              overwrite=True#,
              #restore_from="fresh"
              )
```
