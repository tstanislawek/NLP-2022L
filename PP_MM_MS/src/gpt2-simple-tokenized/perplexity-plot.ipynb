```python
data = []
with open("/home/jupyter/NLP_disco_polo_generator/log_data2.dat", "r") as f:
    text = f.readlines()
    for line in text:
        # 0 means token has been found
        if line.find("Current perplexity = ")==0: 
            data.append(float(line.rsplit(" ")[-1]))
        else:
            pass
```

```python
x = [200*i for i in range(1, len(data)+1)]
plt.plot(x, data)
plt.xlabel("number of steps", fontsize=12)
plt.ylabel("perplexity", fontsize=12)
plt.title("Perplexity on validation set", fontsize=16)
plt.savefig("./perplexity_validation2.png", dpi=400)
plt.show()
```
